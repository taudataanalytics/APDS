# General Form of the Print Function

* print(value1, value2, value3, ..., sep=' ', end='\n', file=sys.stdout, flush=False)
* value = any Python data type
* Reference: https://docs.python.org/3/library/functions.html#print

```python
# OLD Python method
nK = 7
aK = 89.234532
dt = 'Cancer Classification'
print("Number of categories = %d\n with accuracy = %.3f\tData = '%s'" %(nK,aK,dt))

# \n ==> new line .... \t = Tab
# Note: only a space should precede "%(" (not a comma or period)
# %d = digit = integer
# %f = float ... .XYZf ===> number of digits after the decimal point = xyz
# %s = string 
```

```python
# New Python method
nK = 7
aK = 89.234532
dt = 'Cancer Classification'
print("Number of categories = {}\n with accuracy = {}\tData = '{}'".format(nK,aK,dt))
# The ".format()" command is separated by a period
```

```python
# How to format a floating point with this new syntax
print("Number of categories = {}\n with accuracy = {:.2f}\nData = '{}'".format(nK,aK,dt))
# Note this is slightly different from the old syntax for floats... there is now a ":" sign
```

```python
# It can also be unordered
a, b, c = 3, 4.5, 'xyz' # new way of initialization
print('c={2} then a={0}, followed by b={1}'.format(a,b,c))
```

```python
# In the Python print function, we can also specify how variables are separated
print(a,b,c,sep=':')
```

```python
# by default, each call to the print function in Python will create a new line
print(a)
print(b)
print(c)
```

```python
# However, this can be changed
print(a, end=', ')
print(b, end=', ')
print(c, end='.')
```

# Python - Buffer - Flush

* When executing a complex Data Science process in Python, the print output is sent to a buffer and is NOT displayed immediately.
* To force the print output to be displayed immediately, the "flush" parameter must be used (default = False).
* Example:

```python
from time import sleep

for n in range(0,5):
    print("*",end='', flush = True)
    sleep(1.0)
```

## Pointer to Reference

```python
# Be careful with the implicit copy of reference (Pointer to Variable) in Python
A =[2,3,4]
B = A
A.append(7) # We are not doing anything to B
print('A = ', A)
print('B = ', B)

# Why is the output like that???...
# Python avoids "copy of memory" as much as possible, which is suitable for DS/ML
# annoyingly, this kind of behavior sometimes differs for different Python versions.
# Data Scientists need information about changes that occur when using a newer version of Python
# Solution?
```

```python
A =[2,3,4]
B = A.copy() # this is similar to B=A in other programming languages
A.append(7)
print('A = ', A)
print('B = ', B)
```

## Zipping Lists in Data Science
### Processes where two vectors/arrays must be processed together at their corresponding indices are common in DS.

```python
# Zipping List
A = ['Sunday','Monday','Friday', 'Tuesday']
B = ['Holiday', 'Work', 'Work but want a Holiday']

for a,b in zip(A,B):
    print('{} {}'.format(a,b))
    
# What happens if the lists have different sizes???... ==> be careful
# also often used in Parallel Processing in Python ==> discussed in PDS3
```

```python
# Be careful, Python can create a copy of a variable
def my_func(C,D):
    return D+5

A = 10
print(my_func(A, 2))
print(A) # but.... there are still 2 instances of A in RAM
```

```python
# Solution
# Be careful, Python can create a copy of a variable
A = [7, 3, 4]
def my_func(A,B):
    B = A # in Python, B is just a pointer to A .... so there is no memory copy
    # remember the pointer to reference example in the previous list
    return B

print(my_func(A, 2))
print(A) 
```

## ``*args`` and ``**kwargs``: Flexible Arguments

* Sometimes we need to create a function, but we don't know exactly how many parameters will be needed.
* ``*args`` and ``**kwargs`` are used in these situations to capture all the parameters that are passed.
* Function parameters are also sometimes called arguments.
* Here is an example:

```python
def capture_parameters(*args, **kwargs):
    print("type of args = {}, content = {}".format(type(args),args))
    print("type of kwargs = {}, content = {}".format(type(kwargs),kwargs))
```

```python
def summation(*A, **B):
    print(B)
    print(A)
```

```python
X = [3,4,2]
B = 99
summation(X, C=B)
```

```python
capture_parameters(1, 2, 3, a=4, b=5)
```

```python
capture_parameters('a',  keyword=2)
```

```python
# of course, we can also do this
# but be careful, we need the * and ** symbols
T = (1, 2, 3)
D = {'pi': 3.14}

capture_parameters(*T, **D)
```

# Lambdas

* Lambda is a "temporary"/disposable function.
* Often used as an argument/parameter in other functions.
* In other words, there are functions that take a function (lambda) as input.

```python
# Simple example
F = lambda a : a + 10
print(F(5))
```

```python
# Another example
G = lambda a, b, c : a + b * c
print(G(5, 6, 2))
```

## But why not just use a regular function?

```python
# Typically, the use case is like this (Used in banking)
def F(n):
    return lambda a : a * n

P = F(2)
print(P(3))
# understand? ... :) ... 
```

```python
# This simple function sorts a list of tuples based on the second element
# in DS, this kind of thing appears very often
pairs = [(7, 'one'), (2, 'two'), (0, 'three'), (99, 'four'), (5,'afive')] # List of Tuples
pairs.sort(key=lambda p: p[1])  # the name of the lambda function is "key"
pairs
```
