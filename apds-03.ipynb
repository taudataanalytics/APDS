```python
# Add a function for the class average grade

class Student:
    def __init__(self, name, grade):
        self.name = name
        self.grade = grade
        
    def get_grade(self):
        return self.grade

class Course:
    def __init__(self, name, max_students):
        self.name = name
        self.max_students = max_students
        self.students = []
    
    def add_student(self, student):
        if len(self.students) < self.max_students:
            self.students.append(student)
            return True
        else:
            return "Error: Sorry, the class is full"
    def get_average_grade(self):
        sum_ = 0 
        for student in self.students:
            sum_ += student.get_grade()
            # note here we do this because student is an object
            # the student object has the "get_grade" method
        return sum_/len(self.students)
    
s1 = Student('Udin', 77)
s2 = Student('Ucok', 67)
s3 = Student('Asep', 87)

course = Course('Calculus', 2)
course.add_student(s1), course.add_student(s2)
```

# Inheritance in OOP (Python)
* When a *child* class inherits *properties* from a *parent* class, this is called **inheritance**.

# Why use inheritance?

* Code reusability: think of it like a "template".
* Transition & Readability: Good for teamwork.
* Real-world Relationship: Relationships between classes/objects.

```python
# Simplest example of inheritance
class Parent:
    def function1(self):
        print("this is the parent's function")
         
class Child(Parent):
    def function2(self):
        print("this is the child's function")
        
first_born = Child()
# NOTE that "first_born" has the function from "Parent"
first_born.function1(), first_born.function2()
```

```python
# Using __init__ as in the previous lesson 
class Parent:
    def __init__(self, name='Bambang', age='40'):
        self.name = name
        self.age = age
    def function1(self):
        print("this is the parent's function")
    def info(self):# Method from the class as in the previous lesson
        print("Name = {}, Age = {}".format(self.name, self.age))
        
class Child(Parent):
    def __init__(self, name, age, child_number):
        Parent.__init__(self, name, age)
        self.child_number = child_number
    def function2(self):
        print("this is the child's function")
    def info(self):
        print("Name = {}, Age = {}, child number-{}".format(self.name, self.age, self.child_number))
        
first_born = Child("Budi", 5, 2) # "Parent" Property/Method is OVERWRITTEN by "Child"
print(first_born.info())
```

<img alt="" src="images/Types-of-Inheritance.jpg"/>

* **Single**: 1 child, 1 parent (Previous example)
* **Multiple**: More than 1 parent
* **Multilevel**: A child becomes a parent to another child
* **Hierarchical**: A hybrid of the above types.

```python
 # Example of Multiple Inheritance
class Father:
    def __init__(self, name='Bambang', age='40'):
        self.name = name
        self.age = age
    def fatherFunction(self):
        print("this is the Father's function")
    def info(self):
        print("Name = {}, Age = {}".format(self.name, self.age))

class Mother:
    def __init__(self, name='Wati', age='40'):
        self.name = name
        self.age = age
    def motherFunction(self):
        print("this is the Mother's function")
    def info(self):# Method from the class as in the previous lesson
        print("Name = {}, Age = {}".format(self.name, self.age))
        
class Child(Father, Mother):
    def __init__(self, name, age, child_number):
        Father.__init__(self, name, age)
        self.child_number = child_number
    def childFunction(self):
        print("this is the child's function")
    def info(self):
        print("Name = {}, Age = {}, child number-{}".format(self.name, self.age, self.child_number))
        
first_born = Child("Budi", 5, 2) # "Father" & "Mother" properties/methods are inherited by "Child"
print(first_born.fatherFunction(), first_born.motherFunction())
```

```python
 # Example of Multilevel Inheritance
class Grandfather:
    def __init__(self, name='Iwan', age='40'):
        self.name = name
        self.age = age
    def grandfatherFunction(self):
        print("this is the Grandfather's function")
    def info(self):# Method from the class as in the previous lesson
        print("Name = {}, Age = {}".format(self.name, self.age))

class Parent(Grandfather):
    def __init__(self, name='Parto', age='40'):
        self.name = name
        self.age = age
    def parentFunction(self):
        print("this is the Parent's function")
    def info(self):
        print("Name = {}, Age = {}".format(self.name, self.age))
        
class Child(Parent): 
    def __init__(self, name, age, child_number):
        Parent.__init__(self, name, age)
        self.child_number = child_number
    def childFunction(self):
        print("this is the child's function")
    def info(self):
        print("Name = {}, Age = {}, child number-{}".format(self.name, self.age, self.child_number))
        
first_born = Child("Budi", 5, 2) # "Parent and Grandfather" properties/methods are inherited by "Child"
print(first_born.grandfatherFunction())
```

```python
 # Example of Hierarchical inheritance
class Grandfather:
    def __init__(self, name='Iwan', age='40'):
        self.name = name
        self.age = age
    def grandfatherFunction(self):
        print("this is the Grandfather's function")
    def info(self):
        print("Name = {}, Age = {}".format(self.name, self.age))

class Parent(Grandfather):
    def __init__(self, name='Parto', age='40'):
        self.name = name
        self.age = age
    def parentFunction(self):
        print("this is the Parent's function")
    def info(self):
        print("Name = {}, Age = {}".format(self.name, self.age))

class Uncle():
    def __init__(self, name='Denis', age='32'):
        self.name = name
        self.age = age
    def uncleFunction(self):
        print("this is the Uncle's function")
    def info(self):
        print("Name = {}, Age = {}".format(self.name, self.age))
        
class Child(Uncle, Parent): 
    def __init__(self, name, age, child_number):
        Uncle.__init__(self, name, age)
        self.child_number = child_number
    def childFunction(self):
        print("this is the child's function")
    def info(self):
        print("Name = {}, Age = {}, child number-{}".format(self.name, self.age, self.child_number))
        
first_born = Child("Budi", 5, 2) 
print(first_born.uncleFunction(), first_born.grandfatherFunction())
```

# Method Overriding/Overwriting

* MO changes a function in the Parent class.

Method overriding in Python’s object-oriented programming (OOP) refers to the process by which a subclass provides a new implementation of a method that is already defined in its parent or superclass. When a method in the subclass has the same name, parameters, and return type as that in the superclass, the subclass version takes precedence. This mechanism allows a subclass to tailor or extend the behavior inherited from its parent, ensuring that specialized classes can define more context-appropriate functionality while maintaining the same interface.

Overriding is a core principle supporting polymorphism, enabling objects of different classes to respond differently to the same method call. In practice, the subclass may still access the parent’s version of the method using the super() function, allowing for partial reuse of inherited logic. This design promotes flexibility, extensibility, and code maintainability by encouraging hierarchical refinement rather than duplication of functionality.

```python
class Parent():
    def function(self):
        print("this is the Parent's function")
        
class Child(Parent):
    def function(self): # Note the function names are the same
        print("this is inside the Child's function")

first_born = Child()
first_born.function()
```

# The Super() Function

In Python’s object-oriented programming (OOP) paradigm, the super() function provides a mechanism to access methods and properties of a parent or superclass without explicitly naming it. This built-in function is commonly used to invoke an overridden method from the superclass, thereby promoting code reusability and maintainability. By using super(), a subclass can extend or modify the behavior of its parent class while preserving the parent’s original implementation logic.

The super() function plays a particularly important role in multiple inheritance, where a class inherits from more than one parent. In such cases, super() follows the Method Resolution Order (MRO) — a specific hierarchy defined by Python to determine the order in which classes are searched for methods and attributes. This ensures that each class in the inheritance chain is initialized only once, preventing redundancy and potential conflicts.

Conceptually, super() enhances encapsulation and scalability in class design. It allows developers to modify superclass behavior dynamically and maintain polymorphic relationships between objects. Thus, super() is not merely a syntactic convenience but a fundamental component for achieving clean, modular, and extensible OOP architectures in Python.

```python
# Another example

class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

class Square(Rectangle):
    def __init__(self, length, width):
        super().__init__(length,width)

    def area(self):
        return self.length*self.width

class Cube(Rectangle):
    def __init__(self, length, width, height):
        super().__init__(length,width)
        self.height = height
    def volume(self):
        return self.length*self.width*self.height


square = Square(3, 3)
cube = Cube(3, 3, 3)

print(square.area())
print(cube.volume())
```

<center><h2><strong><font color="blue"> Parent VS Super? </font></strong></h2></center>


```python
class Parent:
    def greet(self):
        print("Hello from Parent")

class Child(Parent):
    def greet(self):
        super().greet()          # Dynamic (recommended)
        # Parent.greet(self)     # Explicit (manual)

child = Child()
child.greet()
```

✅ super().greet() is dynamic — it automatically resolves the correct method according to Method Resolution Order (MRO).

⚠️ Parent.greet(self) is static — it hardcodes the parent name and ignores the MRO.

```python
class A:
    def greet(self): print("A")

class B(A):
    def greet(self):
        super().greet()
        print("B")

class C(A):
    def greet(self):
        super().greet()
        print("C")

class D(B, C):
    def greet(self):
        super().greet()
        print("D")

D.mro()
```

## Multiple Inheritance Compatibility

* super() supports multiple inheritance and respects Python’s MRO, ensuring that each superclass method is called exactly once.
* Explicit parent calls can easily break the MRO and cause methods to be executed multiple times or skipped entirely.

## ✅ Use super() — the modern, recommended practice

*super()* is:
* Dynamic: it follows the Method Resolution Order (MRO) automatically.
* Safe for multiple inheritance: avoids duplicate or skipped method calls.
* Maintainable: you can change class names or reorder inheritance without breaking method chaining.
* Clean and standard: universally recognized by Python developers as the idiomatic style.

## What is Method Resolution Order (MRO)?

Method Resolution Order (MRO) defines the linear order in which base classes are searched when executing a method or attribute lookup.

When you call a method like obj.method(), Python doesn’t just look in the class of obj. Instead, it searches through a sequence of classes, starting from the object’s own class, then its parents, and continues following the MRO until it finds the method.

<center><img alt="" src="images/MRO-POOP.jpg"/></center>

```python
class A:
    def greet(self):
        print("Hello from A")

class B(A):
    def greet(self):
        print("Hello from B")

class C(B):
    pass

C().greet()
```

MRO: C → B → A → object

✅ Python looks for greet in C, then B, then A, and finally in object (the base of all classes).

```python
class A:
    def greet(self): print("A")

class B(A):
    def greet(self): print("B")

class C(A):
    def greet(self): print("C")

class D(B, C):
    pass

print(D.mro())  # View the Method Resolution Order
D().greet()
```

✅ Python first looks in D, then B, then C, and so on

it uses the C3 Linearization algorithm to ensure consistency and avoid duplication.

### Python computes MRO using C3 linearization, which follows three principles:
* Preserve local precedence order — a subclass always checks its direct parents first, in the order they appear.
* Monotonicity — the MRO must not contradict any inheritance order defined by base classes.
* Consistency — every class appears only once in the final order.

```python
print(D.__mro__)
# or
print(D.mro())
```
